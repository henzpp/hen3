```python
# onehot encoding 한 파일
titanic_df = pd.read_pickle('tdf.pkl')
titanic_df.head()
```

```python
from sklearn.model_selection import train_test_split

# 독립 변수, 종속 변수 분리
y_tdf = titanic_df['survived']
X_tdf = titanic_df.drop('survived', axis=1)

# 학습용 데이터와 평가용 데이터를 8:2로 분리
X_train, X_test, y_train, y_test = train_test_split(X_tdf,y_tdf, test_size=0.2, random_state=11)
print(X_train.shape, X_test.shape)
```

```python
# 모델 학습 및 평가

from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier()
rf_model.fit(X_train, y_train)
rf_pred = rf_model.predict(X_test) # 독립변수만 줌 나오면 y_test랑 비교. 평가 가능
accuracy_rf = accuracy_score(y_test,rf_pred).round(2)

lr_model = LogisticRegression()
lr_model.fit(X_train, y_train)
lr_pred = lr_model.predict(X_test)
accuracy_lr = accuracy_score(y_test,lr_pred).round(2)

print(f'rf 정확도 : {accuracy_rf}, lr 정확도 : {accuracy_lr}')
```

#### Age null값 처리 (평균값 대체)

```python
tidf['Initial']=0
for i in tidf:
    tidf['Initial'] = tidf.name.str.extract('([A-Za-z]+)\.')
tidf
```

```python
tidf['Initial'].value_counts()
```

```python
tidf['Initial'].replace(['Rev','Dr','Col','Mlle','Ms','Major','Capt','Sir','Dona','Jonkheer','Countess','Don','Mme','Lady'],
                        ['Other','Mr','Other','Miss','Miss','Mrs','Mr','Mrs','Mrs','Mrs','Mrs','Mr','Miss','Mrs'], inplace = True)
tidf.Initial.value_counts()
```

```python
tidf.groupby('Initial')['age'].mean()
```

```python
tidf.loc[(tidf.age.isnull())&(tidf.Initial=='Master'),'age']=5
tidf.loc[(tidf.age.isnull())&(tidf.Initial=='Miss'),'age']=22
tidf.loc[(tidf.age.isnull())&(tidf.Initial=='Mr'),'age']=32
tidf.loc[(tidf.age.isnull())&(tidf.Initial=='Mrs'),'age']=37
tidf.loc[(tidf.age.isnull())&(tidf.Initial=='Other'),'age']=46

tidf.age.isnull().sum()
```

#### one hot encoding -> embarked

```python
df1 = pd.get_dummies(data = tidf, columns = ['embarked'], prefix = 'Embarked')
df1
```

#### one hot encoding -> sex

```python
df2 = pd.get_dummies(data = df1, columns = ['sex'], prefix='Sex')
df2
```

#### family 컬럼 생성 (sibsp + parch)

```python
df3 = df2.copy()
df3['Family'] = df2['sibsp']+df2['parch']
df3
```

```python
df4 = df3.drop(['name','Initial','sibsp','parch'], axis=1)
df4
```

#### Age 범주화

```python
def agecat(x):
    x = (x//10)
    return x

Age_category = df4.age.apply(agecat)
Age_category
```

```python
df4.insert(1, 'Age_category', Age_category)
df4
```
