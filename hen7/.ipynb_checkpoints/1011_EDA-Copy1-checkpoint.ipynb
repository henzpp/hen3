## 3분기 구매고객

```python
cond1 = (all_df['분기']==3)
cond2 = (all_df['구매년도']==2014)
qua3_14 = all_df.loc[cond1&cond2]
qua3_14
```

```python
gby_quart3 = qua3_14.groupby('고객번호').sum()
gby_quart3
```

## 3분기가 아닌 구매고객(2014)

```python
cond1 = (all_df['분기']!=3)
cond2 = (all_df['구매년도']==2014)
qua124_14 = all_df.loc[cond1&cond2]
qua124_14
```

```python
gby_quart124 = qua124_14.groupby('고객번호').sum()
gby_quart124
```

## 3분기 구매고객 구매증감 여부 파악

```python
# 3분기 구매 증가 고객 (비유의 고객)
# 3분기 - (나머지 분기 구매금액/3) > 0 
```

```python
print((gby_quart3['구매금액']-(gby_quart124['구매금액']/3))>0)
```

```python
cond3 = ((gby_quart3['구매금액']-(gby_quart124['구매금액']/3))>0)
index0 = gby_quart3.loc[cond3==True]
print(index0['구매금액'].count())
print(index0.index)
```

```python
# 3분기 구매 감소 고객 (유의 고객)
# 3분기 - (나머지 분기 구매금액/3) <= 0 
```

```python
cond3 = ((gby_quart3['구매금액']-(gby_quart124['구매금액']/3))>0)
index1 = gby_quart3.loc[cond3==False]
print(index1['구매금액'].count())
print(index1.index)
```
