### 물가상승률 반영 함수


```python
def inflation_rate(df):
    price=0
    
    if df['중분류명'] in ['가공식품','신선식품']:
        price = df['구매금액']*1.016
    elif df['중분류명'] in ['외식']:
        price = df['구매금액']*1.023
    elif df['중분류명'] in ['일상용품', '디지털/가전', '가구/인테리어', '유아용품']:
        price = df['구매금액']*1.026
    elif df['중분류명'] in ['의약품/의료기기']:
        price = df['구매금액']*1.013
    elif df['중분류명'] in ['교육/문화용품']:
        price = df['구매금액']*1.017
    elif df['중분류명'] in ['기타', '명품']:
        price = df['구매금액']*1.027
    elif df['중분류명'] in ['의류','패션잡화']:
        price = df['구매금액']*1.013
    elif df['중분류명'] in ['전문스포츠/레저']:
        price = df['구매금액']*0.995
    else:
        price = df['구매금액']*1.007
    return price
```
