#### 그룹연산
- 복잡한 데이터를 어떤 기준에 다라 여러 그룹으로 나누어서 관찰할 수 있으며 이런 방식으로 분할 처리하는 것을 그룹 연산이라 함.
- 그룹 연산은 데이터를 집계, 반환, 필터하는데 효율적이며 판다스 groupby() 메소드를 사용
- 그룹 객체 만들기(분할)
- 그룹 연산 메소드(적용-결합)

````python
import pandas as pd
import seaborn as sns

titanic = sns.load_dataset('titanic')
titanic.columns                          # 중복되는 자료는 데이터 오염시킴 중복은 하나 제거
````

````python
import pandas as pd
import seaborn as sns

titanic = sns.load_dataset('titanic')
df = titanic.loc[:,['age','sex','class','fare','survived']]

print('승객 수:', len(df))
print(df.head())
````

````python
df.columns
````

````python
# class열을 기준으로 그룹핑(분할)

grouped = df.groupby(['class'])
print(grouped)
list(grouped)
````

````python
# 그룹 객체를 iteration(반복)으로 출력

for key, group in grouped :
    print('* key :', key)
    print('* number :', len(group))
    print(group.head())
````

````python
# 연산 메서드 적용
average = grouped.mean()
print(average)               # survived 생존률
````

````python
stat = grouped.max()
stat                        # 알파벳도 카운트함
````

````python
# Q. 3rd 그룹만을 선택해서 group3 이름으로 저장하고 통계요약표를 출력하세요.
group3 = grouped.get_group('Third')
group3.describe()                             # describe 통계요약표
````

````python
grouped_two = df.groupby(['class', 'sex'])
average_two = grouped_two.mean()
average_two
````

````python
group3f = grouped_two.get_group(('Third','female'))
group3f.head()
````

````python
# 모든 열에 여러 함수를 매핑

agg_all = grouped.agg(['min','max'])
agg_all.head()
````

````python
# 각 열마다 다른 함수를 적용하여 집계

agg_sep = grouped.agg({'fare':['min','max'], 'age':'mean'})
agg_sep.head()
````

````python
# Q. 데이터 개수가 200개 이상인 그룹만을 필터링하여 데이터프레임으로 반환하여 
# grouped_filter 이름으로 저장 후 처음 5개 행을 출력하세요.
````

## pivot
피벗 테이블을 구성하는 4가지 요소(행인덱스, 열인덱스, 데이터 값, 데이터 집계함수)에 적용할 데이터프레임의 열을 각각 지정하여 함수의 인자로 전달

````python
import seaborn as sns
titanic = sns.load_dataset('titanic')
df = titanic.loc[:,['age','sex','class','fare','survived']]
df.head()
````

````python
pdf1 = pd.pivot_table(df,
                     index = 'class',
                     columns = 'sex',
                     values = 'age',
                     aggfunc = 'mean')
pdf1.head()                                # 아하 클래스 행 성별 열 만들고 데이터는 나이
                                           # 그르구 그 나이의 평균
````

````python
pdf1 = pd.pivot_table(df,
                     index = 'class',
                     columns = 'sex',
                     values = 'survived',
                     aggfunc = 'mean')
pdf1.head()    
````

````python
pdf1 = pd.pivot_table(df,
                     index = 'class',
                     columns = 'sex',
                     values = 'fare',
                     aggfunc = 'max')
pdf1.head()    
````

````python
# 행, 열, 값에 사용할 열을 2개 이상 지정 - index, values, aggfunc에 각각 2개 이상 지정

pdf2 = pd.pivot_table(df,
                     index = ['class','sex'],
                     columns = 'survived',
                     values = ['age','fare'],
                     aggfunc=['mean','max'])
pdf2
````

[pipe vs applymap vs apply]  

pipe와 applymap과의 차이는 반환하는 결과가 pipe가 더 다양하다는 것으로 applymap은 데이터프레임 원소에 적용하므로 데이터프레임으로만 리턴되지만 pipe는 사용자 함수의 리턴값에 따라 개별 값, 시리즈, 데이터프레임을 반환할 수 있다.  

현재 가장 많이 쓰이는 메서드는 apply인데 그 이유는 다양한 형태로 값을 입력 받을 수 있고, 반환하는 형태도 폭이 넓기 때문. 결과는 같아보이지만, 시리즈를 이용해 데이터프레임의 행과 열도 이용할 수 있으며 시리즈 그 자체도 함수로 활용할 수 있다. 

만약 데이터프레임에 함수를 적용해야한다면 리턴하는 형태가 많은 pipe 사용도 고려


````python
import pandas as pd
df = pd.DataFrame({
    'Col1':[30,40,50,60],
    'Col2':[23,35,65,45],
    'Col3':[85,87,90,89]
}, index=list('ABCD'))
df
````

````python
df.apply(lambda x:x.sum(),axis=0)
````

````python
df.Col1.apply(lambda x:x+1)
````

````python
df.Col1.map(lambda x:x+1)
````

````python
df.applymap(lambda x:'%.2f'% x)
````

````python
# The pipe() method allows you to apply one or more functions to the DataFrame object.

import pandas as pd

def change_age(x):
  x["age"]=[10, 20, 30]
  return x

data = {
  "name": ["Sally", "Mary", "John"],
  "age": [50, 40, 30]
}

df = pd.DataFrame(data)
df
````

````python
df.pipe(change_age)
````

### 데이터프레임 결합
* 판다스 join() 메소드는 merge() 함수를 기반으로 만들어졌기 때문에 기본 작동 방식이 서로 비슷
* join() 메소드는 두 데이터프레임의 행 인덱스를 기준으로 결합하는 점에서 merge와 차이가 있으나 on=keys 옵션 설정하면 열을 기준으로 결합

````python
data1 = np.arange(1, 21).reshape(4, 5)
data2 = np.arange(11, 31).reshape(5, 4)
df1 = pd.DataFrame(data1, columns=list('abcde'))
df2 = pd.DataFrame(data2, columns=list('dehi'))
print(df1,'\n')
print(df2)
````

````python
df1.join(df2, lsuffix='_a',rsuffix='_b')      #suffix - left, right
````

````python
df1.join(df2, lsuffix='_a',rsuffix='_b',how='inner')    # 디폴트는 inner 교집합
````

````python
df1.join(df2, lsuffix='_a',rsuffix='_b', how='outer')      # outer 합집합
````

````python
df1['ind'] = [1,2,3,4]
df2['ind'] = [1,2,3,4,5]
df1
````

````python
# ind 칼럼을 인덱스로 만드는것

df1.set_index('ind').join(df2.set_index('ind'), lsuffix='_a',rsuffix='_b')
````
