```python
import seaborn as sns
df = sns.load_dataset('titanic')
df.head()
```

```python
df.age
```

```python
df.age.value_counts()                # 나이가 너무 다 달라서 의미가 없음. 20대, 30대, 요론식 아니라
```

```python
# 나이 연령대 별로 구분해주기
```

```python
def get_category(x):
    cat = ''
    if x < 20 :
        cat = 'child'
    elif x < 30 :
        cat = 'Young Adult'
    elif x < 50 :
        cat = 'Adult'
    else :
        cat = 'Elderly'
    return cat

df['age_class'] = df.age.apply(lambda x : get_category(x))
df[['age','age_class']].head()
```

```python
# 소문자를 대문자로 변환
```

```python
def upper_(x):
    return x.upper()
df['sex'] = df.sex.apply(lambda x : upper_(x))
df[['sex']].head()
```
